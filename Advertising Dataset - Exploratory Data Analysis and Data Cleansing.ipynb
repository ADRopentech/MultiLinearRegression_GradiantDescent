```python
adv = adv_raw.drop(labels = [adv_raw.columns[0]], axis = 1)
print(adv)
```

```python
pd.set_option('display.max_rows', None)
print(adv)
```

```python
#Checking for missing values
print(adv_raw.isna().sum())
```

```python
adv_boxplot = adv.boxplot(return_type='axes')
print(adv_boxplot)
```

```python
plt.scatter(adv['TV'],adv['sales'])
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
plt.scatter(adv['radio'],adv['sales'])
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

```python
plt.scatter(adv['newspaper'],adv['sales'])
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()
```

```python
print(adv.corr())
```

```python
ax = sns.heatmap(adv.corr(),
            xticklabels=adv.corr().columns, 
            yticklabels=adv.corr().columns,
            cmap='coolwarm',
            annot=True,
            linewidths=.5,
            fmt='.2f',
            center=0,
            square=True,
            cbar=False)
plt.show()
```
