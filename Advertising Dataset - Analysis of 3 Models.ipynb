```python
#PARTITIONING THE DATASET

import math

#select_rows takes a numpy array and a percentage (between zero and one) and returns the training and test sets as numpy arrays.
#The test set size will be that percentage of the original array.
def select_rows(array, percentage):
    size = len(array)
    rows_to_select = int(math.floor(len(array)*percentage))
    print ("Rows to select: ")
    print (rows_to_select)
    print (type(rows_to_select))
    np.random.shuffle(array)
    print (type(array))
    print(array)
    test_rows = array[:rows_to_select]
    tr = rows_to_select - size
    print("tr = " + str(tr))
    train_rows = array[tr:]
    return train_rows, test_rows
```

```python
#Defining a function to transform a numpy array into an appropriate dataframe.
#The first column of the output dataframe contains lists of feature values.
#The second column contains target values.
def regression_array_to_dataframe (array):
    numberOfDataPoints = len(array)
    numberOfFeatures = len(array[0])-1
    arrayV2 = []
    i = 0
    while i < numberOfDataPoints:
        Features = [None]*(numberOfFeatures+1)
        Features[0] = 1
        j=1
        while j < numberOfFeatures+1:
            Features[j] = array[i][j-1]
            j = j+1
        Target = array[i][numberOfFeatures]
        newLine = [None]*2
        arrayV2.append(newLine)
        arrayV2[i][0] = Features
        arrayV2[i][1] = Target
        i = i+1
    dataframe = pd.DataFrame(data=arrayV2, columns = ['featureListPlusOne', 'target'])
    return dataframe
```

```python
#These functions are the same as their namesakes in the 'Advertising Dataset - Validation of Gradient Descent Algorithm' notebook.
#computationV2Multipoint (defined in that notebook) is not used in this notebook.

def computationV2(featureListTargetDf, eta, numberOfSteps):
    featureListPlusOne = featureListTargetDf['featureListPlusOne']
    numberOfFeaturesPlusOne = len(featureListPlusOne[0])
    thetaValues = [0]*numberOfFeaturesPlusOne
    step = 0
    while step < numberOfSteps:
        theta_iter = 0
        while theta_iter < numberOfFeaturesPlusOne:
            thetaValues[theta_iter] = thetaValues[theta_iter] - eta*calcDiff(featureListTargetDf, thetaValues, theta_iter)
            theta_iter = theta_iter+1
        step = step+1
    return(thetaValues)

def calcDiff(featureListTargetDf, thetaValues, theta_iter):
    featureListPlusOne = featureListTargetDf['featureListPlusOne']
    target = featureListTargetDf['target']
    numberOfFeaturesPlusOne = len(featureListPlusOne[0])
    numberOfDataPoints = len(featureListTargetDf)
    data_point_iterator = 0
    running_total = 0
    while data_point_iterator < numberOfDataPoints:
        featureListPlusOneValues = featureListPlusOne[data_point_iterator]
        targetValue = target[data_point_iterator]
        featureListPlusOneValue = featureListPlusOneValues[theta_iter]
        discrepancy = (estimatedTargetValue(featureListPlusOneValues, thetaValues) - targetValue)*featureListPlusOneValue
        running_total = running_total + discrepancy
        data_point_iterator = data_point_iterator + 1
    return (2/numberOfDataPoints)*running_total
                
def estimatedTargetValue(featureListPlusOneValues, thetaValues):
    running_total = 0
    feature_iter = 0
    while feature_iter < len(featureListPlusOneValues):
        running_total = running_total + thetaValues[feature_iter]*featureListPlusOneValues[feature_iter]
        feature_iter = feature_iter+1
    return running_total

def goodnessOfFit(featureListTargetDf, thetaValues):
    featureListPlusOne = featureListTargetDf['featureListPlusOne']
    target = featureListTargetDf['target']
    numberOfFeaturesPlusOne = len(featureListPlusOne[0])
    numberOfDataPoints = len(featureListTargetDf)
    sumOfSquares = 0
    data_point_iterator = 0
    while data_point_iterator < numberOfDataPoints:
        featureListPlusOneValues = featureListPlusOne[data_point_iterator]
        targetValue = target[data_point_iterator]
        discrepancy = (estimatedTargetValue(featureListPlusOneValues, thetaValues) - targetValue)
        sumOfSquares = sumOfSquares + discrepancy**2
        data_point_iterator = data_point_iterator+1
    gOF = 1 - sumOfSquares/((numberOfDataPoints-1)*(target.var()))
    return gOF


    
```

```python
#Making training and test dataframes 
trainDf = regression_array_to_dataframe(train_rows)
print(trainDf)
testDf = regression_array_to_dataframe(test_rows)
print(testDf)
```

```python
#Function for generating thetas and GOF from ETA, steps, trainDf and testDf, putting the results in a dataFrame and printing.
#This is the "shell" used for computationV2. It does not create a trace of intermediate values, but only the final values
#for the thetas and goodness of fit.
def createThetasDf(eta, steps, trainDf, testDf):
    thetas = []
    result = computationV2(trainDf, eta, steps)
    GOF_train = goodnessOfFit(trainDf,result)
    GOF_test = goodnessOfFit(testDf,result)
    result.append(eta)
    result.append(steps)
    result.append(GOF_train)
    result.append(GOF_test)
    thetas.append(result)
    thetasdf = pd.DataFrame(data=thetas,columns=['theta0', 'theta1', 'theta2', 'theta3', 'ETA', 'Steps', 'GOF (training)', 'GOF (test)'])
    print("Thetas = ")
    print(thetasdf)
    print("GOF_train = ")
    print(GOF_train)
    print("GOF_test = ")
    print(GOF_test)
    return thetasdf
```
